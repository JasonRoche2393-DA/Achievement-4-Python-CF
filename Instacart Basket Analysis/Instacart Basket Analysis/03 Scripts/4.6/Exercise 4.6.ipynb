- Both dataframes share the 'product_id' column, meaning a merge is possible here. I will use an Inner Join, as per the explanation on 4.6, "Used to keep only information that’s present in both data sets."

```python
# Test merge without overwriting using the merge flag
pd.merge(orders_products_combined, df_prods, on = ['product_id'] , how = 'inner', indicator = 'match' )
```

7. Export the newly created dataframe as ords_prods_merge in a suitable format

- For a suitable format, I will remove the 'unnamed: 0', '_merge' and 'Unnamed: 0_y' columns from the table.

```python
# Creating df 'ords_prods_merge'
ords_prods_merge = orders_products_combined.merge(df_prods , on = ['product_id'] , indicator = 'match' )
```

```python
# Removing columns
ords_prods_merge.drop(['Unnamed: 0_x', '_merge', 'Unnamed: 0_y'], axis=1)
```
