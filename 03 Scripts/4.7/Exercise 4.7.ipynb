2. Changing "Busiest Day" to "Busiest Days" (plural) and creating a column for the two "Slowest Days"

```python
# Creating "Busiest Days" and "Slowest Days" column
result_2 = []
for value in df_ords_prods_merge["orders_day_of_week"]:
    if value == 0 or value == 1:
        result_2.append("Busiest days")
    elif value == 4 or value == 3: 
        result_2.append("Slowest days")
    else: 
        result_2.append("Regularly busy")
```

4. Create a new column containing these labels called busiest_period_of_day. Use periods of time labeled Most orders, Average orders, and Fewest orders.

```python
# Checking the counts in 'order_hour_of_day' column
df_ords_prods_merge['order_hour_of_day'].value_counts()
```

Using the above information, I have 24 different values. I will use these 24 values to create the labels. My logic for the labels:
"Most orders" = 10, 11, 14, 15, 13, 12, 16, 9
"Average orders" = 17, 8, 18, 19, 20, 7, 21, 22
"Fewest orders" = 23, 6, 0, 1, 5, 2, 4

```python
# Creating a For-Loop If-Statement to label the periods of time, "Most orders", "Average orders" and "Fewest orders"
result_3 = []
for value in df_ords_prods_merge["order_hour_of_day"]:
    if value in [10, 11, 14, 15, 13, 12, 16, 9]: result_3.append("Most orders")
    elif value in [23, 6, 0, 1, 5, 2, 4, 3]: result_3.append("Fewest orders")
    else: result_3.append("Average orders")
```

```python
result_3
```

```python
# Creating a new column for "busiest_period_of_day" within the df ords_prods_merge
df_ords_prods_merge['busiest_period_of_day'] = result_3
```
