4. Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
# Using the transform() function for data aggregation
df['max_order'] = df.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Checking output
df.head(10)
```

```python
# Creating a loyalty flag for existing customers using the loc() function
df.loc[df['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df.loc[(df['max_order'] <= 40) & (df['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df.loc[df['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Printing frequency of the new 'loyalty_flag'
df['loyalty_flag'].value_counts(dropna = False)
```

```python
# Checking output after last execution
df.head(20)
```

5. Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer). What you’re trying to determine is whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers.

6. Create a spending flag for 'Low spender' and 'High spender'

```python
# Using the transform() function for data aggregation
df['avg_product_price'] = df.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Checking output and columns used
df[['user_id', 'avg_product_price']].head(10)
```

```python
# Creating 'Low spender' and 'High spender' flags using loc function
df.loc[df['avg_product_price'] >= 10, 'spending_flag'] = 'High spender'
df.loc[df['avg_product_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# Printing frequency of the new 'low spender' and 'High spender' flags
df['spending_flag'].value_counts(dropna = False)
```

```python
# Using the head.() function to review the output and columns
df[['user_id', 'avg_product_price', 'spending_flag']].head(20)
```

7. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column.

```python
# Using the transform() function for data aggregation
df['median_days_between_orders'] = df.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Using the head.() function to review the output and columns
df[['user_id', 'days_since_prior_order', 'median_days_between_orders']].head(20)
```

```python
# Creating 'Order frequency' flags using loc function
df.loc[df['median_days_between_orders'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
df.loc[(df['median_days_between_orders'] <= 20) & (df['median_days_between_orders'] > 10), 'order_frequency_flag'] = 'Regular customer'
df.loc[df['median_days_between_orders'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# Printing frequency of the 'Order frequency' flags
df['order_frequency_flag'].value_counts(dropna = False)
```

```python
# Using the head.() function to review the output and columns
df[['user_id', 'days_since_prior_order', 'median_days_between_orders', 'order_frequency_flag']].head(20)
```
